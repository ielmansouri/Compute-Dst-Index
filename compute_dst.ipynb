```python
def compute_dst (data_train): 
    data_train_new = data_train.copy()
    W2 = np.array([ 0.2535, 0.3172, -0.4860, -0.2346])
    B2 = np.array([0.0712])
    B1 = np.array([-0.1081, -0.7252, -0.2307, 0.5406])
    Wc = np.array([[0.6099, 0.2223, -0.4379,0.0053],[0.7935, 0.5000, -1.2654, -0.0428],[-0.3496, 0.1024, 0.3605, 0.1790], [-0.2067, 0.2270, 0.5446, 0.1019]])
    W1 = np.array([[-0.5946, 0.0904, 0.0428], [0.6215, 0.4485, 0.2516], [-1.4217, 0.4988, 0.1409],[0.3372, -0.5065, 0.7190]])

    Bz = np.array(data_train_new['Bz'].tolist())
    Np = np.array(data_train_new['Np'].tolist())
    V = np.array(data_train_new['V'].tolist())
    N = len(Bz)
    uc0 = np.array([0.39,0.22,-0.83,0.14])

    u = np.zeros((3,N))
    u[0,:] = Bz/30
    u[1,:] = Np/60.0 - 1.0
    u[2,:] = V/400.0 - 1.5

    y = np.zeros((1,N))
    uc = uc0
    for i in range(N):
        x = np.tanh(W1.dot(u[:,i]) + Wc.dot(uc) + B1)
        y[:,i] = W2.dot(x) + B2
        uc = x
        if math.isnan(y[:,i]):
            uc = uc0
    Dst = 150.0*y - 100.0
    
    data_train_new['Dst'] = Dst[0]
    
    return data_train_new
    
```
